
# Renaming our Column Headers


```python
data.rename(columns = {"hg/ha_yield": "Yield_Ton/Hectare", "average_rain_fall_mm_per_year": "Rainfall_mm", "pesticides_tonnes": "Pesticides_Tonnes",
                      "avg_temp": "Average_Temp"}, inplace =True)
```

```python
data
```

# Standardize Crop Names By Removing References

```python
data.rename(columns = {"Item": "Crop"}, inplace = True)
```

```python
data["Crop"] = data["Crop"].apply(lambda x: x.replace(", paddy", " "))
```

```python
data["Crop"] = data["Crop"].apply(lambda x: x.replace("and others", " "))
```

# Dropping Columns

```python
data.drop(columns=['Unnamed: 0'], inplace = True)
```

```python

data
```

###  So we are converting hg\ha yield to tonnes for easy analysis


```python
data["Yield_Ton/Hectare"] = data["Yield_Ton/Hectare"] / 10000
```

```python
data["Yield_Ton/Hectare"] = data["Yield_Ton/Hectare"].round(2)
```
